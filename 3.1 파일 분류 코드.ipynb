##태림산업 비전불량 추가 이미지 이미지 파일 분류

```python
import os
import shutil

# 기본 경로 설정
source_path = '/content/drive/MyDrive/data/태림산업 비전불량 추가이미지'
destination_base_path = '/content/drive/MyDrive/data/태림산업 비전불량 추가이미지'

# 이미지 파일을 각도 및 Shaft/Tube별로 정리
for image in os.listdir(source_path):
    # 파일 경로 설정
    image_path = os.path.join(source_path, image)

    # 파일인지 확인
    if os.path.isfile(image_path):
        # 파일명을 '_'로 분할하여 정보 추출
        parts = image.split('_')
        if len(parts) >= 4:
            part_type = parts[2]  # Shaft 또는 Tube
            angle = parts[3]      # 0도, 90도, 180도, 270도 중 하나

            # 폴더 이름 생성
            folder_name = f"{part_type}_{angle}"
            folder_path = os.path.join(destination_base_path, folder_name)
            os.makedirs(folder_path, exist_ok=True)

            # 이미지 파일을 새로운 폴더로 이동
            destination_path = os.path.join(folder_path, image)
            shutil.move(image_path, destination_path)
            print(f"Moved {image} to {destination_path}")

print("이미지 정리가 완료되었습니다.")
```

## 태림산업 비전불량 이미지 SHAFT 분류

```python
import os
import shutil

# 기본 경로 설정
base_path = '/content/drive/MyDrive/data/태림산업 비전불량 이미지/SHAFT'

# 각 하위 폴더 탐색
for subfolder in os.listdir(base_path):
    subfolder_path = os.path.join(base_path, subfolder)

    if os.path.isdir(subfolder_path):
        # 하위 폴더 이름을 사용하여 파일명에 추가
        subfolder_parts = subfolder.split(' ')
        folder_suffix = '_'.join(subfolder_parts) if subfolder_parts else ''

        # 하위 폴더 내의 파일 탐색
        for file in os.listdir(subfolder_path):
            file_path = os.path.join(subfolder_path, file)

            if os.path.isfile(file_path):
                # 이미지 파일 처리
                if file.startswith('Shaft_') and file.endswith('.bmp'):
                    # 파일 이름에서 각도 추출
                    _, angle = file.split('_', 1)
                    angle_folder = os.path.join(base_path, f"Shaft_{angle}")

                    # 각도 폴더 생성
                    os.makedirs(angle_folder, exist_ok=True)

                    # 새로운 파일 이름 생성
                    new_file_name = f"{os.path.splitext(file)[0]}_{folder_suffix}{os.path.splitext(file)[1]}"
                    destination_path = os.path.join(angle_folder, new_file_name)

                    # 파일 이동
                    shutil.move(file_path, destination_path)
                    print(f"Moved {file_path} to {destination_path}")

                # Result.csv 파일 처리
                elif file == 'Result.csv':
                    result_folder = os.path.join(base_path, 'Result.csv')
                    os.makedirs(result_folder, exist_ok=True)

                    # 새로운 파일 이름 생성
                    new_file_name = f"Result_{folder_suffix}.csv"
                    destination_path = os.path.join(result_folder, new_file_name)

                    # 파일 이동
                    shutil.move(file_path, destination_path)
                    print(f"Moved {file_path} to {destination_path}")

        # 하위 폴더가 비어 있으면 삭제
        if not os.listdir(subfolder_path):
            os.rmdir(subfolder_path)
            print(f"Removed empty folder: {subfolder_path}")

print("이미지 및 CSV 파일이 분류되었고, 빈 하위 폴더가 삭제되었습니다.")
```

## 태림산업 비전합격 이미지 파일 분류

```python
import os
import shutil

# 기본 경로 설정
base_path = '/content/drive/MyDrive/data/태림산업 합격 이미지'
shaft_folder = os.path.join(base_path, 'Shaft')
tube_folder = os.path.join(base_path, 'Tube')
csv_folder = os.path.join(base_path, 'CSV')

# Shaft, Tube 및 CSV 폴더 생성
os.makedirs(shaft_folder, exist_ok=True)
os.makedirs(tube_folder, exist_ok=True)
os.makedirs(csv_folder, exist_ok=True)

# 상위 폴더 내의 모든 하위 폴더 탐색
for subfolder in os.listdir(base_path):
    subfolder_path = os.path.join(base_path, subfolder)

    if os.path.isdir(subfolder_path):
        # 하위 폴더 이름에서 정보를 추출
        parts = subfolder.split(' ')
        second_last_part = parts[-2] if len(parts) > 1 else ''

        if second_last_part.startswith('S'):
            # Shaft로 분류
            target_folder = shaft_folder
        elif second_last_part.startswith('T'):
            # Tube로 분류
            target_folder = tube_folder
        else:
            continue  # S나 T로 시작하지 않으면 무시

        # 하위 폴더 내의 모든 파일을 대상 폴더로 이동
        for file in os.listdir(subfolder_path):
            file_path = os.path.join(subfolder_path, file)
            if os.path.isfile(file_path):
                # 이미지 파일 처리
                if file.startswith('Shaft_') or file.startswith('Tube_'):
                    # 파일 이름에서 각도 추출
                    if '_' in file:
                        prefix, angle = file.split('_', 1)
                        angle_folder = os.path.join(target_folder, f"{prefix.lower()}_{angle}")

                        # 각도 폴더 생성
                        os.makedirs(angle_folder, exist_ok=True)

                        # 새로운 파일 이름 생성
                        new_file_name = f"{os.path.splitext(file)[0]}_{'_'.join(parts)}{os.path.splitext(file)[1]}"
                        destination_path = os.path.join(angle_folder, new_file_name)

                        # 파일 이동
                        shutil.move(file_path, destination_path)
                        print(f"Moved {file_path} to {destination_path}")

                # CSV 파일 처리
                elif file == 'Result.csv':
                    # 새로운 파일 이름 생성
                    new_file_name = f"Result_{'_'.join(parts)}.csv"
                    destination_path = os.path.join(csv_folder, new_file_name)

                    # 파일 이동
                    shutil.move(file_path, destination_path)
                    print(f"Moved {file_path} to {destination_path}")

        # 하위 폴더가 비어 있으면 삭제
        if not os.listdir(subfolder_path):
            os.rmdir(subfolder_path)
            print(f"Removed empty folder: {subfolder_path}")

print("모든 파일이 적절한 폴더로 이동되었고, 빈 하위 폴더가 삭제되었습니다.")
```

## 비전불량 이미지 - SHAFT데이터 개수

```python
# 비전 불량 이미지 SHAFT
import os

# 기본 디렉토리 정의
base_dir = '/content/drive/MyDrive/data/태림산업 비전불량 이미지/SHAFT'

# 파일 수를 세는 함수
def count_files_in_directory(directory):
    file_count = 0
    for subdir, _, files in os.walk(directory):
        file_count += len(files)
    return file_count

# 파일 수 계산 및 출력
total_files = count_files_in_directory(base_dir)
print(f"총 파일 수: {total_files}")
```
